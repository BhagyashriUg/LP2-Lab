```python
import pandas as pd
from sklearn.datasets import load_iris

# Load Iris dataset (offline, no file needed)
iris = load_iris()

df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Convert target to species names
df['species'] = iris.target
df['species'] = df['species'].map({
    0: 'Iris-setosa',
    1: 'Iris-versicolor',
    2: 'Iris-virginica'
})

# Clean column names (like your original code)
df.columns = df.columns.str.strip().str.lower()

df.head()
```

```python
setosa_stats = df[df['species'] == 'Iris-setosa'].describe()
versicolor_stats = df[df['species'] == 'Iris-versicolor'].describe()
virginica_stats = df[df['species'] == 'Iris-virginica'].describe()

print("Setosa Statistics:\n", setosa_stats)
print("\nVersicolor Statistics:\n", versicolor_stats)
print("\nVirginica Statistics:\n", virginica_stats)
```

```python
import pandas as pd
from sklearn.datasets import load_iris

# Load dataset (offline)
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add species column
df['species'] = iris.target
df['species'] = df['species'].map({
    0: 'Iris-setosa',
    1: 'Iris-versicolor',
    2: 'Iris-virginica'
})

# Clean column names
df.columns = df.columns.str.strip().str.lower()

# 🔥 IMPORTANT LINE (this is what assignment wants)
grouped_stats = df.groupby('species').describe()

print(grouped_stats)
```
