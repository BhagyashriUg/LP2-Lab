```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
```

```python
# Dataset Information

print("Dataset Name: Iris Dataset")
print("Source: UCI Machine Learning Repository")
print("URL: https://archive.ics.uci.edu/ml/datasets/iris")
```

```python
# Load Iris dataset
iris = sns.load_dataset('iris')

# Display first 5 rows
iris.head()
```

```python
# Shape of dataset

print("Shape of Dataset:", iris.shape)
```

```python
# Column names

print("Columns in Dataset:")
print(iris.columns)
```

```python
# Data types of each column

print("Data Types:")
print(iris.dtypes)
```

```python
# Statistical summary

iris.describe()
```

```python
# Check missing values

print("Missing Values:")
print(iris.isnull().sum())
```

```python
# Variable descriptions

print("""
Feature Description:

1. sepal_length  -> Length of sepal (numeric)
2. sepal_width   -> Width of sepal (numeric)
3. petal_length  -> Length of petal (numeric)
4. petal_width   -> Width of petal (numeric)
5. species       -> Type of iris flower (categorical)
""")
```

```python
# Histogram for each feature

iris.hist(figsize=(10,8), bins=15)

plt.tight_layout()
plt.show()
```

```python
# Boxplots for all numeric features

plt.figure(figsize=(12,8))

# Sepal Length
plt.subplot(2,2,1)
sns.boxplot(y=iris['sepal_length'])
plt.title('Sepal Length')

# Sepal Width
plt.subplot(2,2,2)
sns.boxplot(y=iris['sepal_width'])
plt.title('Sepal Width')

# Petal Length
plt.subplot(2,2,3)
sns.boxplot(y=iris['petal_length'])
plt.title('Petal Length')

# Petal Width
plt.subplot(2,2,4)
sns.boxplot(y=iris['petal_width'])
plt.title('Petal Width')

plt.tight_layout()
plt.show()
```

```python
# Check data types before conversion

print("Before Conversion:")
print(iris.dtypes)

# Convert data type example
iris['sepal_length'] = iris['sepal_length'].astype(float)

print("\nAfter Conversion:")
print(iris.dtypes)
```

```python
# Normalize numeric columns

scaler = MinMaxScaler()

numeric_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

iris_normalized = iris.copy()

iris_normalized[numeric_columns] = scaler.fit_transform(
    iris_normalized[numeric_columns]
)

print("Normalized Dataset:")
iris_normalized.head()
```

```python
# Label Encoding for species column

label_encoder = preprocessing.LabelEncoder()

iris['species_label'] = label_encoder.fit_transform(iris['species'])

print(iris[['species', 'species_label']].head())
```

```python
# One Hot Encoding

one_hot_encoded = pd.get_dummies(iris, columns=['species'])

one_hot_encoded.head()
```

```python
# Correlation heatmap

plt.figure(figsize=(8,6))

sns.heatmap(
    iris.corr(numeric_only=True),
    annot=True,
    cmap='YlGnBu'
)

plt.title('Correlation Heatmap')

plt.show()
```

```python
# Pairplot

sns.pairplot(iris, hue='species')

plt.show()
```
